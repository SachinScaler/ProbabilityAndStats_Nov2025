### Outliers by hand

```python
import numpy as np

data = [25, 28, 28, 30, 33, 36, 37, 37, 100, 38, 39, 41, 44, 47]
data.sort()
```

```python
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
```

```python
iqr = q3 - q1
iqr
```

```python
p_min =  q1 - 1.5 * iqr
p_max = q3 + 1.5 * iqr
p_min, p_max
```

```python
outliers = list(filter(lambda x: x < p_min or x > p_max, data))
outliers
```

A data analyst is trying to convince their team that customer engagement is stable. The monthly active users (MAU) for the last 10 months are: [18, 22, 22, 25, 27, 28, 29, 30, 31, 95]

The team argues that the data is highly skewed, but the analyst responds:
"Well, the mean and median aren't too far apart, and the mode is one of the most common values, so clearly engagement is stable."
Based on mean, median, mode, standard deviation, and range,

Which of the following statements most exposes the flaw in the analystâ€™s argument?

```python
from scipy import stats
data = np.array([18, 22, 22, 25, 27, 28, 29, 30, 31, 95])

print("Mean", np.mean(data))
print("Median", np.median(data))
print("Mode", stats.mode(data))
print("Standard Deviation", np.std(data))
print("Range", np.max(data) - np.min(data))
```
